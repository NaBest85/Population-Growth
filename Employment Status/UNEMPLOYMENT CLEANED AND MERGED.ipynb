Function to read, clean, and rename the 'US States' column in the data


```python
def read_df(filename):
    df = pd.read_csv(filename)
    return df

def clean_rename_data(df):
    
    df = df.dropna()  # Example: Remove rows with missing values


    # Rename 'US States' column to 'Name'
    df.rename(columns={'US States': 'Name'}, inplace=True)

    return df
```

Read into pandas df seperately 

```python
# Read, clean, and rename data for each year
df_2020 = read_df(path_to_2020_csv)
df_2021 = read_df(path_to_2021_csv)
df_2022 = read_df(path_to_2022_csv)
```

Clean the data 

```python
df_2020 = clean_rename_data(df_2020)
df_2021 = clean_rename_data(df_2021)
df_2022 = clean_rename_data(df_2022)

```
